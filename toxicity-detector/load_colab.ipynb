```python
# Define a function to clean text while keeping obfuscated symbols
def clean_text(text):
    text = text.lower()
    text = re.sub(r'<.*?>', '', text)  # Remove HTML
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Remove URLs
    text = re.sub(r'[^a-z0-9\s@#$%&*!?]', '', text)  # Keep common obfuscation symbols
    text = re.sub(r'\s+', ' ', text).strip()  # Normalize spaces
    return text

# Apply cleaning
train_df['clean_text'] = train_df['comment_text'].astype(str).apply(clean_text)
for i in range(3):
    print(f"Original: {train_df['comment_text'][i]}")
    print(f"Cleaned:  {train_df['clean_text'][i]}")
    print("-" * 80)
train_df[['comment_text', 'clean_text']].head()
```

Tokenization Helper and PyTorch Dataset

```python
# Tokenization helper function
def tokenize_text(texts, tokenizer, max_len=128):
    return tokenizer(
        list(texts),
        padding='max_length',
        truncation=True,
        max_length=max_len,
        return_tensors="pt"
    )

class ToxicDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=128):
        self.texts = list(texts)
        self.labels = labels.values
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        # Tokenize one sample at a time (faster and memory-safe)
        encoding = self.tokenizer(
            self.texts[idx],
            truncation=True,
            padding='max_length',
            max_length=self.max_len,
            return_tensors='pt'
        )
        item = {key: val.squeeze(0) for key, val in encoding.items()}  # remove batch dim
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item

# # Custom PyTorch Dataset
# class ToxicDataset(Dataset):
#     def __init__(self, texts, labels, tokenizer, max_len=128):
#         self.encodings = tokenize_text(texts, tokenizer, max_len)
#         self.labels = torch.tensor(labels.values, dtype=torch.long)

#     def __len__(self):
#         return len(self.labels)

#     def __getitem__(self, idx):
#         item = {key: val[idx] for key, val in self.encodings.items()}
#         item['labels'] = self.labels[idx]
#         return item
```

Step 5: Model development and training

```python
import torch
import torch.nn as nn
from transformers import BertForSequenceClassification, get_scheduler
from torch.optim import AdamW
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import os
import time
```

Load BERT + Binary Classification Head

```python
import torch
from transformers import BertForSequenceClassification

# Check for GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

try:
    # Load BERT model
    model = BertForSequenceClassification.from_pretrained(
        "bert-base-uncased",
        num_labels=2  # for binary classification
    )
    model.to(device)

    if torch.cuda.is_available():
        print(f"âœ… BERT model loaded and moved to device: {torch.cuda.get_device_name(0)}")
    else:
        print("âš ï¸ CUDA not available. Model loaded on CPU.")
except Exception as e:
    print("âŒ Error loading BERT model:", e)
```

Set Optimizer, Loss, Scheduler

```python
# Set up optimizer
optimizer = AdamW(model.parameters(), lr=2e-5, weight_decay=0.01)
print("âœ… Optimizer (AdamW) initialized with learning rate 2e-5 and weight decay 0.01.")

# Set number of epochs
epochs = 4
print(f"ðŸ“† Number of training epochs: {epochs}")

# Calculate total training steps
train_steps = len(train_loader) * epochs
print(f"ðŸ” Total training steps: {train_steps} ({len(train_loader)} batches x {epochs} epochs)")

# Set up learning rate scheduler
scheduler = get_scheduler(
    name="linear",
    optimizer=optimizer,
    num_warmup_steps=0,
    num_training_steps=train_steps
)
print("âœ… Learning rate scheduler (linear) set up successfully.")

# Loss function
loss_fn = nn.CrossEntropyLoss()
print("âœ… Loss function (CrossEntropyLoss) initialized.")
```

 Helper: Compute Metrics

```python
def compute_metrics(preds, labels):
    preds = np.argmax(preds, axis=1)
    accuracy = accuracy_score(labels, preds)
    precision = precision_score(labels, preds)
    recall = recall_score(labels, preds)
    f1 = f1_score(labels, preds)
    return accuracy, precision, recall, f1
```

 Training + Validation Function

```python
def train_one_epoch(model, loader, optimizer, scheduler):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    print("ðŸš€ Starting training epoch...")
    for step, batch in enumerate(loader):
        try:
            optimizer.zero_grad()
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            logits = outputs.logits

            loss.backward()
            optimizer.step()
            scheduler.step()

            total_loss += loss.item()
            all_preds.extend(logits.detach().cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

            if step % 100 == 0:
                print(f"ðŸŸ¡ Step {step}/{len(loader)} - Loss: {loss.item():.4f}")

        except Exception as e:
            print(f"âŒ Error in training batch at step {step}:", e)

    avg_loss = total_loss / len(loader)
    acc, prec, rec, f1 = compute_metrics(all_preds, all_labels)
    print("âœ… Training epoch complete.")
    print(f"ðŸ“Š Train Metrics â†’ Loss: {avg_loss:.4f} | Acc: {acc:.4f} | Prec: {prec:.4f} | Rec: {rec:.4f} | F1: {f1:.4f}")
    return avg_loss, acc, prec, rec, f1


def evaluate(model, loader):
    model.eval()
    total_loss = 0
    all_preds, all_labels = [], []

    print("ðŸ” Starting evaluation...")
    with torch.no_grad():
        for step, batch in enumerate(loader):
            try:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)

                outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
                loss = outputs.loss
                logits = outputs.logits

                total_loss += loss.item()
                all_preds.extend(logits.detach().cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

                if step % 100 == 0:
                    print(f"ðŸ”¹ Eval Step {step}/{len(loader)} - Loss: {loss.item():.4f}")

            except Exception as e:
                print(f"âŒ Error in evaluation batch at step {step}:", e)

    avg_loss = total_loss / len(loader)
    acc, prec, rec, f1 = compute_metrics(all_preds, all_labels)
    print("âœ… Evaluation complete.")
    print(f"ðŸ“Š Val Metrics â†’ Loss: {avg_loss:.4f} | Acc: {acc:.4f} | Prec: {prec:.4f} | Rec: {rec:.4f} | F1: {f1:.4f}")
    return avg_loss, acc, prec, rec, f1
```

Train Loop with Early Stopping + Logging + Checkpoint

```python
import os
import time
import torch

# Create checkpoints directory if it doesn't exist
os.makedirs("checkpoints", exist_ok=True)

# Early stopping setup
best_val_f1 = 0
patience = 2
wait = 0

# Training loop
for epoch in range(epochs):
    print(f"\nðŸ” Epoch {epoch + 1}/{epochs}")

    start_time = time.time()

    # Train for one epoch
    train_loss, train_acc, train_prec, train_rec, train_f1 = train_one_epoch(
        model, train_loader, optimizer, scheduler
    )

    # Validate after epoch
    val_loss, val_acc, val_prec, val_rec, val_f1 = evaluate(model, val_loader)

    # Print metrics
    print(f"ðŸŸ© Train | Loss: {train_loss:.4f}, Acc: {train_acc:.4f}, Prec: {train_prec:.4f}, Rec: {train_rec:.4f}, F1: {train_f1:.4f}")
    print(f"ðŸŸ¦ Val   | Loss: {val_loss:.4f}, Acc: {val_acc:.4f}, Prec: {val_prec:.4f}, Rec: {val_rec:.4f}, F1: {val_f1:.4f}")

    # Save the model if validation F1 improved
    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        wait = 0
        torch.save(model, "checkpoints/best_model.pkl")
        print("âœ… Model improved. Saved as checkpoints/best_model.pkl.")
    else:
        wait += 1
        if wait >= patience:
            print("â›” Early stopping triggered.")
            break

    duration = time.time() - start_time
    print(f"â±ï¸ Epoch duration: {duration:.2f} seconds")
```

SHAP Explanation Engine (GPU-Compatible)

```python
import shap
import torch
from transformers import BertTokenizer, BertForSequenceClassification, pipeline
import matplotlib.pyplot as plt

# Set device to GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Step 1: Load model and tokenizer
try:
    tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
    model = torch.load("checkpoints/best_model.pkl", map_location=device)
    model.to(device)
    model.eval()
    print("Model and tokenizer loaded successfully.")
except Exception as e:
    print(" Error loading model or tokenizer:", e)

# Step 2: Create Hugging Face pipeline for SHAP (using GPU)
try:
    classifier = pipeline(
        "text-classification",
        model=model,
        tokenizer=tokenizer,
        return_all_scores=True,
        device=0  # GPU
    )
    print("Hugging Face pipeline created successfully (using GPU).")
except Exception as e:
    print(" Error creating pipeline:", e)

# Step 3: Sample texts (you can change/add more)
examples = {
    "clean": "I hope you have a great day!",
    "ambiguous": "That guy is so gay, like really.",
    "adversarial": "Go k1ll urself you id10t"
}

# Step 4: Run SHAP explanations
try:
    print("\nðŸ” Initializing SHAP explainer...")
    explainer = shap.Explainer(classifier)

    for label, text in examples.items():
        print(f"\n Explaining: {label.upper()} sample")
        print(f" Input: {text}")
        shap_values = explainer([text])
        print(f" SHAP explanation tokens (importance):")
        shap.plots.text(shap_values[0])
except Exception as e:
    print(" SHAP explanation error:", e)
```
